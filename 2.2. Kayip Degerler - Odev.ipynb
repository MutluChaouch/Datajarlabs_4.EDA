```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Veri kümesinde yer alan değişkenlerinin tiplerini ve her bir değişkenin eksik (null) değer oranını bulun.

states = pd.read_csv('states_all.csv')
states.info()
states.isnull().sum()/states.shape[0]*100
```

```python
states.head()
```

```python
# Verimizde yıl (year) sütunu olduğunu farketmişsinizdir. Şimdilik yıl verisini unutun ve her bir gözlemin aynı 
# yıl içerisinde yapıldığını farz edin. Her bir değişken için eksik değerleri nasıl doldurabileceğinizi düşünün. 
# Eksik değerleri bir değerle doldurmak hangi değişkenler için anlamlı, hangileri için anlamsızdır?



# Bazi degiskenlerde eksik deger orani cok yuksek (ozellikle avg_math, avg_reading). Verinin turune ve yapacagimiz 
# analize gore belirli bir kabul limiti secilebilir. Bu soru icin GRADES_ALL_G sutunu uzerinden sorulari cevaplayacagim.

states['GRADES_ALL_G_N']=states['GRADES_ALL_G'].fillna(states['GRADES_ALL_G'].mean())
```

```python
# Şimdi zaman faktörünü dikkate alma zamanı! 2. sorudaki cevabınızı tekrar gözden geçirin ve eksik verileri o yıl 
# içerisinde gözlemlenen değerlere dayanarak doldurun. 
# Örneğin, bir değeri ortalama değer ile doldurmak isterseniz, o yılın ortalamasını hesaplayın.

states['GRADES_ALL_G_Y'] = states['GRADES_ALL_G'].fillna(states.groupby('YEAR')['GRADES_ALL_G'].transform('mean'))
```

```python
# Bu sefer, eksik değerleri enterpolasyon yaparak doldurun.

states['GRADES_ALL_G_INT'] = states['GRADES_ALL_G'].interpolate()
```

```python
# İkinci, üçüncü ve dördüncü sorulardaki cevaplarınızı karşılaştırın. Anlamlı bir fark görebiliyor musunuz?

states[['GRADES_ALL_G_N','GRADES_ALL_G_Y','GRADES_ALL_G_INT']].describe()

# Asagidaki kodlarda da gorecegimiz uzere her bir yil icin ortalama deger alip kayip degerleri doldurdugumuz yontemde
# halen daha kayip degerlerimiz oldugunu goruyoruz. Bunun sebebi de 2019 yilinda hic bir mevcut veri olmamasi. 

# Bunun disinda, asagidaki 1. ve 3. sutunlarda gorecegimiz gibi, iki farkli yontemde ortalama ve standart sapma 
# degerlerinde farkliliklar oldu.
```
